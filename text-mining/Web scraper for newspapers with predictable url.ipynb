```python
#create list of urls, the only difference being the date expressed on the url

from datetime import date, timedelta

la_razon_url=[]
startdate = date(int(2012), int(1), int(1))
while startdate < date(int(2016), int(5), int(17)):
    la_razon_url.append('http://www.la-razon.com/portadas-anteriores/'+startdate.strftime('%Y%m%d'))
    startdate += timedelta(days=1)

```

```python
#declare utf-8 encoding for encoding spanish special characters
# -*- coding: utf-8 -*-

#import modules

# for establishing a connection between python and the targeted website
import urllib2

# for retreiving information from websites 
from bs4 import BeautifulSoup as bs 

#To use numpy arrays which are more efficient than regular data structures
import numpy as np

#to add randomised waiting time
import time, random
```

```python
#function to extract content from each url. The content is that belong each day's main headline news.

def get_html_content(url):

    #first access the homepage 
    html = urllib2.urlopen(url).read()
    soup = bs(html)
    
    #extract link pointing to the main headline news
    h2_list=soup.find_all(name='h2')

    for h2 in h2_list:
        try:
            if h2['class'] == ['headline', 'headline-regular']:
                main_headline = h2
        except:
            None
    
    for link in main_headline.find_all('a'):
        main_headline_url='http://www.la-razon.com/'+link.get('href')

    #access the main headline news
    headline_html = urllib2.urlopen(main_headline_url).read()
    headline_soup = bs(headline_html)
    
    headline_raw_text=headline_soup.findAll('p', { "class" : "mce" })
    
    text=''
    for x in headline_raw_text:
        text=text + x.get_text()
    
    text=text.strip()
    
    #paste date string with headline news content encoded in utf-8
    text=url[-8:] + " " + text.encode("utf-8")
        
    return(text)
```

```python
#create function to scrape data from a list of urls

def scrape_contracts(html_urls, file_prefix):
    
    project_info = []

    for idx, url in enumerate(html_urls):
         #random delay between 1 and 2 secs
        delay = random.randint(1,2)
        #pause execution for delay seconds
        time.sleep(delay)
        print 'file: ' + file_prefix + str(idx) + ' delay= ' + str(delay) 
        
        try:
            #try to retrieve information from url
            project_info.append(get_html_content(url))
            
            #add completed url to the log of completed urls
            with open("./completed_urls.txt", "a") as complete_file:
                complete_file.write(url + '\n')
                complete_file.close()
        except:
            #add rejected urls to the log of rejected urls
            with open("./rejected_urls.txt", "a") as rejected_file:
                rejected_file.write(url + '\n')
                rejected_file.close()
        
        if idx % 1000 == 0 and idx != 0:
            #periodically write the data to file and reinitialise list for memory management
            file_name = file_prefix + str(idx) + '.gz'
            np.savetxt(file_name, project_info, delimiter=',', fmt='%s')
            project_info = []
            
            #add the index of last file to be written to disk
            with open("./saved_data_index.txt", "a") as saved_file:
                saved_file.write(file_prefix + str(idx) + '\n')
                saved_file.close()
    
    #Save remaining data to file    
    file_name = file_prefix + str(idx) + '.gz'
    np.savetxt(file_name, project_info, delimiter=',', fmt='%s')
    with open("./saved_data_index.txt", "a") as saved_file:
                saved_file.write(file_prefix + str(idx) + '\n')
                saved_file.close()
```

```python
scrape_contracts(la_razon_url, 'LEO')
```
